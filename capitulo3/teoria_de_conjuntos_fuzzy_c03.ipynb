## Função Triangular

\begin{equation}\label{ftriangular}
triangle(x:a,b,c)=\left\{\begin{matrix}
0, & x<a & \\ 
\frac{x-a}{b-a}, & a \leq x < b & \\ 
\frac{c-x}{c-b}, & b \leq  x < c& \\ 
0, & x > c & 
\end{matrix}\right.
\end{equation}

### Código Python da Funçao de Inclusão Triangular

```python
def triangle(x,a,b,c):
    r = 0
    if a > x:
        r = 0;
    elif a <= x and x < b:
        r = (x-a)/(b-a)
    elif b <= x and x < c:
        r = (c-x)/(c-b)
    elif x > c:
        r = 0
    return r
```

```python
x3 = np.arange(0,10,0.1)
v3 = [];
for i in x3:
    a = triangle(i,2,5,8)
    v3.append(a)
print(len(x3))
```

```python
# grafico
plt.plot(x3,v3,'r', linewidth=1.5, label='Função Triangular')
plt.fill(x3,v3,'r',alpha=0.2)
plt.xlabel('Universo de discurso')
plt.ylabel('valor de pertinência')
plt.legend()
plt.grid(True)
plt.savefig('triangular.png');
#plt.show()
```

## Função Trapezoidal

\begin{equation}\label{ftrapezoidal} 
trapezoid(x:a,b,c,b)=\left\{\begin{matrix}
0, & x < a & \\ 
\frac{x-a}{b-a}, & a \leq x < b & \\ 
1, & b \leq x < c & \\ 
\frac{d-x}{d-c}, & c \leq  x < d & \\ 
0, & x \geq d & 
\end{matrix}\right.
\end{equation}

### Código da Função de Inclusão Trapezoidal

```python
def trapezoid(x,a,b,c,d):
    r = 0
    if a > x:
        r = 0;
    elif a <= x and x < b:
        r = (x-a)/(b-a)
    elif b <= x and x < c:
        r = 1
    elif c <= x and x < d:
        r = (d-x)/(d-c)
    elif x >= d:
        r = 0
    return r
```

```python
x4 = np.arange(0,10,0.1)
v4 = [];
for i in x4:
    a = trapezoid(i,1,3,5,8)
    v4.append(a)
print(len(v4))
```

```python
# grafico
plt.plot(x4,v4,'r', linewidth=1.5, label='Função Trapezoidal')
plt.fill(x4,v4,'r',alpha=0.2)
plt.xlabel('Universo de discurso')
plt.ylabel('valor de pertinência')
plt.legend()
plt.grid(True)
plt.savefig('trapezoidal.png');
#plt.show()
```

## Função Gaussiana

\begin{equation}\label{fgaussiana}
gaussian(x:m,\sigma)=exp\left (- \frac{(x-m)^{2}}{2\sigma^{2}} \right )
\end{equation}

### Código da Função de Inclusão Gaussiana

```python
def gaussian(x,m,s):
    r = 0
    ex = (x-m)*(x-m)
    s2 = s*s
    valor = ex/s2
    va = -0.5*valor
    r = np.exp(va)
    return r
```

```python
xg = np.arange(-5,15,0.1)
g = [];
for i in xg:
    a = gaussian(i,5,2)
    g.append(a)
print(len(xg)) # numero de elementos
```

```python
# grafico
plt.plot(xg,g,'r', linewidth=1.5, label='Função Gaussiana')
plt.fill(xg,g,'r',alpha=0.2)
plt.xlabel('Universo de discurso')
plt.ylabel('valor de pertinência')
plt.legend()
plt.grid(True)
plt.savefig('gaussiana.png');
#plt.show()
```

## Função Sigmoidal

\begin{equation}\label{fsigmoidal}
sigmoid(x:a,c)=\frac{1}{1+exp[{-a(x-c)}]} 
\end{equation}

### Código da Função de Inclusão Sigmoidal

```python
# função sigmoidal
import math
def sigmoid(x,a,c):
    val = -a*(x-c)
    return 1 / (1 + math.exp(val))
```

```python
# gerando o dominio da funçõa sigmoidal
xs = np.arange(0,10,0.1)
s = [];
for i in xs:
    v = sigmoid(i,2,5)
    s.append(v)
#print(len(xs)) # imprime os valores
#print(len(s))
```

```python
# grafico da função sigmoidal
plt.plot(xs,s,'r', linewidth=1.5, label='Função Sigmoidal')
#plt.fill(xs,s,'r',alpha=0.2)
plt.xlabel('Universo de discurso')
plt.ylabel('valor de pertinência')
plt.legend()
plt.grid(True)
plt.savefig('sigmoidal.png');
#plt.show()
```

## Função de bell

\begin{equation}\label{fbell}
bell(x:a,b,c)=\frac{1}{1+|{\frac{x-c}{a}}|^{2b}} 
\end{equation}

### Código da Função de Inclusão Bell

```python
# função bell
def bell(x,a,b,c):
    f = np.power(np.absolute((x-c)/a),2*b)
    return 1/(1+f)
```

```python
# gerando o dominio para a função bell
xb = np.arange(1,100)
b = [];
for i in xb:
    s = bell(i,20,4,50)
    b.append(s)
#print(len(b)) # imprime os valores
```

```python
# grafico da função bell
plt.plot(xb,b,'r', linewidth=1.5, label='Função Bell')
plt.fill(xb,b,'r',alpha=0.2)
plt.xlabel('Universo de discurso')
plt.ylabel('valor de pertinência')
plt.legend()
plt.grid(True)
plt.savefig('bell.png');
```

# Gráfico da função Bell

```python
# gerando o dominio para a função bell
xb = np.arange(1,100)
b = [];
for i in xb:
    s = bell(i,20,4,50)
    b.append(s)
#print(len(b)) # imprime os valores
```

```python
# gerar valores da função bell
def bell_val(l,a,b,c):
    yb = []
    for i in l:
        xx = bell(i,a,b,c)
        yb.append(xx)
    return yb
```

```python
# gerando os valores para bell(x,a,b,c)
xb = np.arange(1,100)
# a = 20
yb1 = bell_val(xb,20,4,50)
# a = 15
yb2 = bell_val(xb,15,4,50)
# a = 10
yb3 = bell_val(xb,10,4,50)
```

```python
#fig, bell = plt.subplots(nrows=1, ncols=1, figsize=(10, 8))
plt.plot(xb,yb1,'r', linewidth=1.5, label='$a=20$')
plt.plot(xb,yb2,'b', linewidth=1.5, label='$a=15$')
plt.plot(xb,yb3,'g', linewidth=1.5, label='$a=10$')
plt.fill(xb,yb3,'g', alpha=0.2)
plt.grid(True)
plt.title('Mudando $a$')
plt.ylabel('valores de pertinência')
plt.legend()
plt.show()
```

## Subconjunto Fuzzy

A é um **subconjunto** de B se e somente se $\mu_{A}(x) \leq \mu_{B}(x)$ para todo $x$. Em símbolos, $ A \subset B \iff  \mu_{A}(x) \leq \mu_{B}(x)$
