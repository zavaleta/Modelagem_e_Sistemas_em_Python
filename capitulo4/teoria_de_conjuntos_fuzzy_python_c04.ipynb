```python
# importando as bibliotecas necessarias
#
# importando a biblioteca para operações numericas no python
import numpy as np
# importando a biblioteca fuzzy
import skfuzzy as fuzzy
# importando a biblioteca para os gráficos
import matplotlib.pyplot as plt
```

```python
# definição do conjunto universo
X = np.arange(0,10.2,0.2)
print('X:',X)
```

## Funções de Pertinência

### Função Triangular

```python
# definido a função triangular (trimf) com valores a=0,b=5,c=10
mf3 = fuzzy.trimf(X,[0,5,10])
print('mf3 =',mf3)
```

```python
# graficando a função de pertinencia triangular
fig, mf3x = plt.subplots(nrows=1, figsize=(8, 5))
mf3x.plot(X,mf3,'r', linewidth=1.5, label='Triangular')
mf3x.fill(X,mf3,'r',alpha=0.2)
mf3x.set_xlabel('Universo de discurso X')
mf3x.set_ylabel('Valores de pertinência')
mf3x.set_title('$trimf(X,[0,5,10])$')
mf3x.legend();
# gravando a figura
fig.savefig('mf3.png');
```

### Função Trapezoidal

```python
# definindo a função trapezoidal (trapmf) com valores a=0,b=3,c=7,d=10
mf4 = fuzzy.trapmf(X,[0,3,7,10])
print('mf4 =',mf4)
```

```python
# graficando a função de pertinencia trapezoidal
fig, mf4x = plt.subplots(nrows=1, figsize=(8, 5))
mf4x.plot(X,mf4,'r', linewidth=1.5, label='Trapezoidal')
mf4x.fill(X,mf4,'r',alpha=0.2)
mf4x.set_xlabel('Universo de discurso X')
mf4x.set_ylabel('Valores de pertinência')
mf4x.set_title('$trapmf(X,[0,3,7,10])$')
mf4x.legend();
# gravando a figura
fig.savefig('mf4.png');
```

### Função de Pertinência Gaussiana

```python
# definir a função gaussiana (gaussmf) com valores c= 5 (centro), sigma=1.5 (largura)
mfg = fuzzy.gaussmf(X,5,1.5)
print('mfg =',mfg)
```

```python
# graficando a função de pertinencia gaussiana
fig, mfgx = plt.subplots(nrows=1, figsize=(8, 5))
mfgx.plot(X,mfg,'r', linewidth=1.5, label='Gaussiana')
mfgx.fill(X,mfg,'r',alpha=0.2)
mfgx.set_xlabel('Universo de discurso X')
mfgx.set_ylabel('Valores de pertinência')
mfgx.set_title('$gaussmf(X,5,1,5)$')
mfgx.legend();
# gravando a figura
fig.savefig('mfg.png');
```

### Função de Pertinência Sigmoidal

```python
# definir a função sigmoidal (sigmf) com valores a=5 (centro) e b=4 (inclinação)
mfs = fuzzy.sigmf(X,5,4)
print('mfs =',mfs)
```

```python
# graficando a função de pertinencia sigmoidal
fig, y_mfs = plt.subplots(nrows=1, figsize=(8, 5))
y_mfs.plot(X,mfs,'r', linewidth=1.5, label='Sigmoidal')
#y_mfs.fill(X,mfs,'r',alpha=0.2)
y_mfs.set_xlabel('Universo de discurso X')
y_mfs.set_ylabel('Valores de pertinência')
y_mfs.set_title('$sigmf(X,5,4)$')
y_mfs.legend();
# gravando a figura
fig.savefig('mfs.png');
```

### Função de Pertinência Bell

```python
# definir a função bell (gbellmf)
# largura equidistante ao centro c no valor de pertinência y=1.0, a = 2
a = 2
# largura ao redor do centro c, b = 1.5
b = 1.5 
# posição central da figura c=5
c = 5 
mfb = fuzzy.gbellmf(X, a, b, c)
print('mfb =',mfb)
```

```python
# graficando a função de pertinencia sigmoidal
fig, y_mfb = plt.subplots(nrows=1, figsize=(8, 5))
y_mfb.plot(X,mfb,'r', linewidth=1.5, label='Bell(2,1,5,5)')
y_mfb.fill(X,mfb,'r',alpha=0.2)
y_mfb.set_xlabel('Universo de discurso X')
y_mfb.set_ylabel('Valores de pertinência')
y_mfb.set_title('$gbellmf(X,2,1.5,5)$')
y_mfb.legend();
# gravando a figura
fig.savefig('mfb.png');
```

## Operações com Conjuntos Fuzzy

### Subconjuntos Fuzzy

```python
# definir os conjuntos fuzzy
# funcao de pertinencia do conjunto A
mu_A = fuzzy.trimf(X,[0,5,10])
# função de pertinencia do conjunto B
mu_B = fuzzy.trapmf(X,[0,3,7,10])
print('A:',np.around(mu_A, decimals=2))
print('B:',np.around(mu_B, decimals=2))
```

```python
# funcao para calcular um subconjunto fuzzy
def subcF(muA, muB):
    d1 = len(muA)
    d2 = len(muB)
    out = []
    if (len(muA)==len(muB)):
        i=0
        while(i < len(muB)):
            x = np.around(muA[i],decimals=2)
            y = np.around(muB[i],decimals=2)
            if x <= y:
                out.append(x)
            else:
                out=[]
            i +=1
    else:
        print('vetores com dimensões diferentes')
    return out
```

```python
# subconjunto A <= B
sub= subcF(mu_A,mu_B)
print('sub:',sub)
```

```python
# graficando a função de pertinencia bell
fig, y_mfsub = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
# grafico na posição 0 (primeira linha e coluna)
y_mfsub[0].plot(X,mu_A,'r', linewidth=1.5, label='$\mu_{A}(x)$')
#y_mfsub[0].fill(X,mu_A,'r',alpha=0.2)
y_mfsub[0].plot(X,mu_B,'b', linewidth=1.5, label='$\mu_{B}(x)$')
#y_mfsub[0].fill(X,mu_B,'b',alpha=0.2)
y_mfsub[0].set_xlabel('Universo de discurso X')
y_mfsub[0].set_ylabel('Valores de pertinência')
y_mfsub[0].set_title('(a) $\mu_{A}(x)$, $\mu_{B}(x)$')
y_mfsub[0].legend();
# grafico na segunda coluna (primeira linha e segunda coluna)
y_mfsub[1].plot(X,mu_A,'r--', linewidth=1.5, label='$\mu_{A}(x)$')
#y_mfsub[1].fill(X,mu_A,'r',alpha=0.2)
y_mfsub[1].plot(X,mu_B,'b--', linewidth=1.5, label='$\mu_{B}(x)$')
#y_mfsub[1].fill(X,mu_B,'b',alpha=0.2)
y_mfsub[1].plot(X,sub,'r', linewidth=1.5, label='$\mu_{A}(x) \leq \mu_{B}(x)$')
y_mfsub[1].fill(X,sub,'r',alpha=0.2)
y_mfsub[1].set_xlabel('Universo de discurso X')
y_mfsub[1].set_ylabel('Valores de pertinência')
y_mfsub[1].set_title('(b) $\mu_{A}(x) \leq \mu_{B}(x)$')
y_mfsub[1].legend();
# gravando a figura
fig.savefig('mfsub.png');
```

### União de Conjuntos Fuzzy

```python
# definir os conjuntos fuzzy
# funcao de pertinência do conjunto A
mu_A4 = fuzzy.trapmf(X,[0,2,4,6])
# funcao de pertinência do conjunto B
mu_B4 = fuzzy.trapmf(X,[4,6,8,10])
```

```python
# funcao que calcula a uniao de dois conjuntos fuzzy
def unionF(muA,muB):
    out = np.fmax(muA,muB)
    return out
```

```python
# saida da uniao de conjuntos fuzzy
unionAB = unionF(mu_A4, mu_B4)
print('union:',unionAB)
```

```python
# graficando a uniao de conjuntos fuzzy
fig, y_mfu = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
# grafico na posição 0 (primeira linha e coluna)
y_mfu[0].plot(X,mu_A4,'g', linewidth=1.5, label='$\mu_{A}(x)$')
y_mfu[0].plot(X,mu_B4,'b', linewidth=1.5, label='$\mu_{B}(x)$')
y_mfu[0].set_xlabel('Universo de discurso X')
y_mfu[0].set_ylabel('Valores de pertinência')
y_mfu[0].set_title('(a) $\mu_{A}(x)$, $\mu_{B}(x)$')
y_mfu[0].legend();
# grafico na segunda coluna (primeira linha e segunda coluna)
y_mfu[1].plot(X,mu_A4,'r--', linewidth=1.5, label='$\mu_{A}(x)$')
y_mfu[1].plot(X,mu_B4,'b--', linewidth=1.5, label='$\mu_{B}(x)$')
y_mfu[1].plot(X,unionAB,'r', linewidth=1.5, label='$max(\mu_{A}(x),\mu_{B}(x))$')
y_mfu[1].fill(X,unionAB,'r',alpha=0.2)
y_mfu[1].set_xlabel('Universo de discurso X')
y_mfu[1].set_ylabel('Valores de pertinência')
y_mfu[1].set_title('(b) $max(\mu_{A}(x),\mu_{B}(x))$')
y_mfu[1].legend();
# gravando a figura
fig.savefig('mfu.png');
```

### Interseção de Conjuntos Fuzzy

```python
# definir os conjuntos fuzzy
# funcao de pertinência do conjunto A
#mu_A4 = fuzzy.trapmf(X,[0,2,4,6])
# funcao de pertinência do conjunto B
#mu_B4 = fuzzy.trapmf(X,[4,6,8,10])
```

```python
# funcao que calcula a intersecao de dos conjuntos fuzzy
def intersectionF(muA,muB):
    out = np.fmin(muA,muB)
    return out
```

```python
# Calculando a intersecao de dois conjuntos fuzzy
intAB = intersectionF(mu_A4,mu_B4)
print('intAB:',intAB)
```

```python
# graficando a intersecao de conjuntos fuzzy
fig, y_mfi = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
# grafico na posição 0 (primeira linha e coluna)
y_mfi[0].plot(X,mu_A4,'g', linewidth=1.5, label='$\mu_{A}(x)$')
y_mfi[0].plot(X,mu_B4,'b', linewidth=1.5, label='$\mu_{B}(x)$')
y_mfi[0].set_xlabel('Universo de discurso X')
y_mfi[0].set_ylabel('Valores de pertinência')
y_mfi[0].set_title('(a) $\mu_{A}(x)$, $\mu_{B}(x)$')
y_mfi[0].legend();
# grafico na segunda coluna (primeira linha e segunda coluna)
y_mfi[1].plot(X,mu_A4,'g--', linewidth=1.5, label='$\mu_{A}(x)$')
y_mfi[1].plot(X,mu_B4,'b--', linewidth=1.5, label='$\mu_{B}(x)$')
y_mfi[1].plot(X,intAB,'r', linewidth=1.5, label='$min(\mu_{A}(x),\mu_{B}(x))$')
y_mfi[1].fill(X,intAB,'r',alpha=0.2)
y_mfi[1].set_xlabel('Universo de discurso X')
y_mfi[1].set_ylabel('Valores de pertinência')
y_mfi[1].set_title('(b) $min(\mu_{A}(x)$, $\mu_{B}(x))$')
y_mfi[1].legend();
# gravando a figura
fig.savefig('mfi.png');
```

### Complemento de Conjuntos Fuzzy

```python
# determinando o complemento de um conjunto Fuzzy
cA = fuzzy.fuzzy_not(mu_A)
print('cA:',cA)
```

```python
# Graficando o complemento de um conjuntos fuzzy
fig, coA = plt.subplots(nrows=1, figsize=(8, 5))
#
coA.plot(X,mu_A,'r', linewidth=1.5, label='$\mu_{A}(x)$')
coA.fill(X,mu_A,'r',alpha=0.2)
coA.legend()
#
coA.plot(X,cA,'-.b', linewidth=1.5, label='$1-\mu_{A}(x)$')
coA.fill(X,cA,'b',alpha=0.2)
coA.set_ylabel('Valores de pertinência')
coA.set_xlabel('Universo de discurso $X$')
coA.set_title('$\mu_{A}(x)$ e $1-\mu_{A}(x)$')
coA.legend()
fig.savefig('ca.png');
```
