## Métodos para Números Fuzzy Triangulares

```python
# Calcula os intervalos para numeros fuzzy triangulares
# usando alfa
def int_tri_alfa(a,m,b,alfa):
    out = []
    if (alfa >= 0) and (alfa <=1):
        i = (m-a)*alfa+a
        s = (m-b)*alfa+b
        out = [i,s]
    else:
        out=[]
        print('Ops!!  0 <= alfa <= 1 !!! ')
    return out
```

```python
# Aplicacao de funcioamento intervalo triangular
I = int_tri_alfa(1,3,5,0.5)
print('Intervalo número fuzzy triangular com alfa=0.5:')
print(I)
```

```python
# # Calcula os intervalos para numeros fuzzy triangulares
# sem alfa
def int_triangular(a,m,b):
    out = []
    alfas = np.arange(0,1.1,0.1)
    if (a != []) and(m != []) and (b != []):
        for alfa in alfas:
            i = np.round((m-a)*alfa+a,2)
            s = np.round((m-b)*alfa+b,2)
            out.append([i,s])
    else:
        out=[]
        print('Ops!! falta um parâmetro!!')
    return out
```

```python
# Apliacacao de funcionamento numero fuzzy triangular
I1 = int_triangular(1,3,5)
print('Intervalos do número fuzzy triangular:')
print(I1)
```

## Métodos para Números Fuzzy Trapezodais

```python
# Calcula os intervalos para numeros fuzzy trapezoidais
# usando alfa
def int_trap_alfa(a,m,n,b,alfa):
    out = []
    if (alfa >= 0) and (alfa <=1):
        i = np.round((m-a)*alfa+a,2)
        s = np.round((n-b)*alfa+b,2)
        out = [i,s]
    else:
        out=[]
        print('Ops!!  0 <= alfa <= 1 !!! ')
    return out
```

```python
# Aplicacao do metodo trapezoidal
T1 = int_trap_alfa(11,14,17,20,0.5)
print('Intervalo do número fuzzy trapezoidal com alfa=0.5:')
print(T1)
```

```python
# # Calcula os intervalos para numeros fuzzy trapezoidais
# sem alfa
def int_trapezoidal(a,m,n,b):
    out = []
    alfas = np.arange(0,1.1,0.1)
    if (a != []) and(m != []) and (n != []) and(b != []):
        for alfa in alfas:
            i = np.round((m-a)*alfa+a,2)
            s = np.round((n-b)*alfa+b,2)
            out.append([i,s])
    else:
        out=[]
        print('Ops!! falta um parâmetro!!')
    return out
```

## Adição de Intervalos

```python
# Adicao de 2 Intervalos
def add_int(A,B):
    la = len(A)
    lb = len(B)
    out = []
    if (la == lb) and (la == 2):
        a_i = A[0]
        a_s = A[1]
        b_i = B[0]
        b_s = B[1]
        i = np.round(a_i + b_i,2)
        s = np.round(a_s + b_s,2)
        out =[i,s]
    else:
        out =[]
        print('ops!! Intervalos de dimensões diferentes!!')
    return out
```

```python
# Aplicacao da adicao de intervalos
A = [1,5]
B = [3,7]
ai = add_int(A,B)
print('Adição de Intervalos:')
print(ai)
```

## Subtração de Intervalos

```python
# Subtracao de Intervalos
def subt_int(A,B):
    la = len(A)
    lb = len(B)
    out = []
    if (la == lb) and (la == 2):
        a_i = A[0]
        a_s = A[1]
        b_i = B[0]
        b_s = B[1]
        i = np.round(a_i - b_s,2)
        s = np.round(a_s - b_i,2)
        out =[i,s]
    else:
        out =[]
        print('ops!! Intervalos de dimensões diferentes!!')
    return out
```

```python
# Aplicacao da subtracao de intervalos
A = [1,5]
B = [3,7]
si = subt_int(A,B)
print('Subtração de Intervalos:')
print(si)
```

## Multiplicação de Intervalos

```python
# Multiplicacao de Intervalos
def mult_int(A,B):
    la = len(A)
    lb = len(B)
    out = []
    if (la == lb) and (la == 2):
        a_i = A[0]
        a_s = A[1]
        b_i = B[0]
        b_s = B[1]
        v = [a_i*b_i, a_i*b_s, a_s*b_i, a_s*b_s]
        i = np.min(v)
        s = np.max(v)
        out =[i,s]
    else:
        out =[]
        print('ops!! Intervalos de dimensões diferentes!!')
    return out
```

```python
# Aplicacao da multiplicacao de intervalos
A = [1,5]
B = [3,7]
mi = mult_int(A,B)
print('Multiplicação de Intervalos:')
print(mi)
```

## Divisão de Intervalos

```python
# Divisao de Intervalos
def div_int(A,B):
    la = len(A)
    lb = len(B)
    out = []
    if (la == lb) and (la == 2):
        a_i = A[0]
        a_s = A[1]
        b_i = B[0]
        b_s = B[1]
        if (b_i != 0) and (b_s != 0):
            dB = [1/b_s,1/b_i]
            m = mult_int(A,dB)
            out = m
        else:
            out = []
            print('Ops!! Divisão por B=[0,0]!!!')
    else:
        out =[]
        print('ops!! Intervalos de dimensões diferentes!!')
    return out
```

```python
# Aplicacao da divisao de intervalos
A = [-1,2]
B = [5,6]
di = div_int(A,B)
print('Divisão de Intervalos:')
print(di)
```

## Máximo de intervalos

```python
# Maximo de Intervalos
def max_int(A,B):
    la = len(A)
    lb = len(B)
    out = []
    if (la == lb) and (la == 2):
        a_i = A[0]
        a_s = A[1]
        b_i = B[0]
        b_s = B[1]
        i = np.max([a_i,b_i])
        s = np.max([a_s,b_s])
        out = [i,s]
    else:
        out =[]
        print('ops!! Intervalos de dimensões diferentes!!')
    return out
```

# Operações com Números Fuzzy Triangulares

```python
# determina numero fuzzy triangular
def num_ftri(v):
    l = len(v)
    out = []
    a = np.round(v[0][0],2) # alfa = 0
    b = np.round(v[0][1],2) # alfa = 0
    m = np.round(v[l-1][0],2) # alfa = 1
    out = [a,m,b]
    return out
```

## Adição de Números Fuzzy Triangulares

```python
# Adicao de numeros fuzzy triangulares
def add_nf_triangulares(A,B):
    la = len(A)
    lb = len(B)
    out = []
    soma = []
    if (la == lb) and (la == 3):
        a1 = A[0]
        m1 = A[1]
        b1 = A[2]
        a2 = B[0]
        m2 = B[1]
        b2 = B[2]
        iA = int_triangular(a1,m1,b1)
        iB = int_triangular(a2,m2,b2)
        for ia,ib in zip(iA,iB):
            s_ab = add_int(ia,ib)
            soma.append(s_ab)
        #out = soma # mostra todos os valores
        out = num_ftri(soma) # numero fuzzy final
    return out
```

```python
# Aplicacao de adicao de numeros fuzzy triangulares
A = [1,3,5]
B = [3,5,7]
A3 = add_nf_triangulares(A,B)
print('Adição de números fuzzy triangulares:')
print('C =',A,' + ',B,' =',A3)
```

```python
# Soma lista de Intervalos
def sum_list_intervals(I1,I2):
    l1 = len(I1)
    l2 = len(I2)
    out = []
    if (l1 == l2) and (l1 > 1):
        for i,j in zip(I1,I2):
            s = add_int(i,j)
            out.append(s)
    elif (l1 == 1):
        out = add_int(I1[0],I2[0])
    else:
        out =[]
        print('ops!! Listas de intervalos de dimensões diferentes!!!')
    return out
```
